```python
## make a copy of the csv file to create a checkpoint 
arabic_corpus = pd.DataFrame(arabic)
french_corpus = pd.DataFrame(french)
turkish_corpus = pd.DataFrame(turkish)
```

```python
# clean he data
french_corpus = french_corpus.drop(index=range(39, len(french_corpus)))
turkish_corpus = turkish_corpus.drop(index=range(39, len(turkish_corpus)))
arabic_corpus = arabic_corpus.drop(index=range(39, len(arabic_corpus)))
```

```python
### azure
arabic_corpus["azure"] = arabic_corpus["english"].apply(lambda x: azure.translate_text_azure(x, 'ar'))
french_corpus["azure"] = french_corpus["english"].apply(lambda x: azure.translate_text_azure(x, 'fr'))
turkish_corpus["azure"] = turkish_corpus["english"].apply(lambda x: azure.translate_text_azure(x, 'tr'))
```

```python
### deepl 
arabic_corpus["deepl"] = arabic_corpus["english"].apply(lambda x: deepl.translate_text_deepl(x, 'ar'))
french_corpus["deepl"] = french_corpus["english"].apply(lambda x: deepl.translate_text_deepl(x, 'fr'))
turkish_corpus["deepl"] = turkish_corpus["english"].apply(lambda x: deepl.translate_text_deepl(x, 'tr'))
```

```python
### gpt-4
arabic_corpus["gpt-4"] = arabic_corpus["english"].apply(lambda x: gpt_4.translate_text(x, 'ar'))
french_corpus["gpt-4"] = french_corpus["english"].apply(lambda x: gpt_4.translate_text(x, 'fr'))
turkish_corpus["gpt-4"] = turkish_corpus["english"].apply(lambda x: gpt_4.translate_text(x, 'tr'))
```

```python
### gpt-4
arabic_corpus["gpt-4o-mini"] = arabic_corpus["english"].apply(lambda x: gpt_4o_mini.translate_text(x, 'ar'))
french_corpus["gpt-4o-mini"] = french_corpus["english"].apply(lambda x: gpt_4o_mini.translate_text(x, 'fr'))
turkish_corpus["gpt-4o-mini"] = turkish_corpus["english"].apply(lambda x: gpt_4o_mini.translate_text(x, 'tr'))
```

```python
### gpt-4
arabic_corpus["gpt-4o"] = arabic_corpus["english"].apply(lambda x: gpt_4o.translate_text(x, 'ar'))
french_corpus["gpt-4o"] = french_corpus["english"].apply(lambda x: gpt_4o.translate_text(x, 'fr'))
turkish_corpus["gpt-4o"] = turkish_corpus["english"].apply(lambda x: gpt_4o.translate_text(x, 'tr'))
```

```python
# evaluation
import sacrebleu

data = turkish_corpus

# Initialize a results dictionary to store BLEU and chrF scores for each model
results = {
    "Model": [],
    "BLEU": [],
    "chrF": []
}

# Get reference translations and convert them to a list format compatible with sacrebleu
references = [[ref] for ref in data['refrences'].tolist()]  # List of lists for sacrebleu compatibility

# Evaluate each model column
for model in data.columns:
    if model != 'refrences' and model != 'english':  # Skip the reference column itself
        translations = data[model].tolist()
        
        # Calculate BLEU and chrF scores
        bleu_score = sacrebleu.corpus_bleu(translations, references).score
        chrf_score = sacrebleu.corpus_chrf(translations, references).score
        
        # Store the results
        results["Model"].append(model)
        results["BLEU"].append(bleu_score)
        results["chrF"].append(chrf_score)

# Create a DataFrame to display the summary table
summary_df = pd.DataFrame(results)
print(summary_df)
```
